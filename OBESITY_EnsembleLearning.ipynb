```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(
    obesity_df.drop(['NObeyesdad','id'],axis=1),
    obesity_df['NObeyesdad'],
    test_size=0.2,#20% data for testing
    random_state=42,
    stratify=obesity_df['NObeyesdad'])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

solver=['lbfgs','liblinear','newton-cg','newton-cholesky','sag','saga']
best_solver=''
test_score=np.zeros(6)
for i,n in enumerate(solver):
  lr = LogisticRegression(solver=n).fit(X_train,y_train)
  test_score[i]=lr.score(X_test,y_test)
  if lr.score(X_test,y_test)==test_score.max():
    best_solver=n
print(best_solver)
lr=LogisticRegression(solver=best_solver)
lr.fit(X_train,y_train)
lr_pred=lr.predict(X_test)
print(accuracy_score(y_test,lr_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

dtree=DecisionTreeClassifier(class_weight='balanced')
param_grid={
    'max_depth':[3,4,5,6,7,8],
    'min_samples_split':[2,3,4],
    'min_samples_leaf':[1,2,3,4],
    'random_state':[0,42]
}

grid_search=GridSearchCV(dtree,param_grid,cv=5)
grid_search.fit(X_train,y_train)
Ctree= DecisionTreeClassifier(**grid_search.best_params_, class_weight='balanced')
Ctree.fit(X_train,y_train)
dtc_pred=Ctree.predict(X_test)
print("DecisionTrees 's Accuracy: ", accuracy_score(y_test,dtc_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rfc=RandomForestClassifier()
param_grid={
    'n_estimators':[50,100,150,200],
    'max_features':['sqrt','log2',None],
    'max_depth':[3,6,9,19],
    'max_leaf_nodes':[3,6,9],
}
grid_search=GridSearchCV(rfc,param_grid)
grid_search.fit(X_train,y_train)
rfctree=RandomForestClassifier(**grid_search.best_params_)
rfctree.fit(X_train,y_train)
rfc_pred=rfctree.predict(X_test)
print("RandomForestClassifier 's Accuracy: ", accuracy_score(y_test,rfc_pred))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import f1_score

kernels={'linear':0, 'poly':0,'rbf':0,'sigmoid':0}
best=''
for i in kernels:
  svm= SVC(kernel=i)
  svm.fit(X_train,y_train)
  yhat=svm.predict(X_test)
  kernels[i]=f1_score(y_test,yhat,average="weighted")
  if kernels[i]== max(kernels.values()):
    best=i
print(best)
svm=SVC(kernel=best)
svm.fit(X_train,y_train)
svm_pred=svm.predict(X_test)
print(f'Svm kernel({best}): {f1_score(y_test,svm_pred,average="weighted")}')
```

WEIGHTED VOTING


```python
from sklearn.ensemble import VotingClassifier
lr    = LogisticRegression()
svm   = SVC(probability=True)
Ctree = DecisionTreeClassifier()

# â”€â”€ HARD VOTING â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
voting_hard = VotingClassifier(
    estimators=[('lr',   lr),       # id string 'lr'
                ('svm',  svm),      # id string 'svm'
                ('tree', Ctree)],   # id string 'tree', object Ctree
    voting='hard'
)
voting_hard.fit(X_train, y_train)
y_pred = voting_hard.predict(X_test)

print("Ensemble acc:", accuracy_score(y_test, y_pred))
# â”€â”€ SOFT VOTING â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
voting_soft = VotingClassifier(
    estimators=[('lr',   lr),
                ('svm',  svm),
                ('tree', Ctree)],
    voting='soft'
)
voting_soft.fit(X_train, y_train)
y_pred = voting_soft.predict(X_test)

print("Ensemble acc:", accuracy_score(y_test, y_pred))
# â”€â”€ WEIGHTED SOFT VOTING â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
voting_weighted = VotingClassifier(
    estimators=[('lr',   lr),
                ('svm',  svm),
                ('tree', Ctree)],
    voting='soft',
    weights=[2, 1, 3]      # order == estimators order
)
voting_weighted.fit(X_train, y_train)
y_pred = voting_weighted.predict(X_test)

print("Ensemble acc:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import BaggingClassifier, VotingClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
# 4) Helper to train/print one bagged model -----------------------------------
def run_bagging(base_estimator, label, n_est=100):
    bag = BaggingClassifier(
        estimator=base_estimator,
        n_estimators=n_est,
        max_samples=0.8,      # 80â€¯% of rows per bag
        max_features=1.0,     # all columns
        bootstrap=True,
        oob_score=True,
        n_jobs=-1,
        random_state=42
    )
    bag.fit(X_train, y_train)
    print(f"{label:>15} | OOB: {bag.oob_score_:.3f} | "
          f"Test: {accuracy_score(y_test, bag.predict(X_test)):.3f}")
    return bag

# 5) Bagged Decision Tree (most variance reduction) ---------------------------
dt_base = DecisionTreeClassifier(random_state=42)
bag_dt = run_bagging(dt_base, "BaggedÂ Tree", n_est=100)

# 6) Bagged Logistic Regression ----------------------------------------------
lr_pipe = make_pipeline(StandardScaler(),
                        LogisticRegression(max_iter=1000, n_jobs=-1, random_state=42))
bag_lr = run_bagging(lr_pipe, "BaggedÂ LogReg", n_est=50)

# 7) Bagged SVM (linear kernel) ----------------------------------------------
svm_pipe = make_pipeline(StandardScaler(),
                         SVC(kernel="linear", probability=False))  # set True if you need proba
bag_svm = run_bagging(svm_pipe, "BaggedÂ SVM", n_est=30)

# 8) (Optional) Voting ensemble on top of the three bagged models -------------
voting = VotingClassifier(
    estimators=[("dt", bag_dt), ("lr", bag_lr), ("svm", bag_svm)],
    voting="soft",           # use "soft" only if *all* models have predict_proba
    n_jobs=-1
)
voting.fit(X_train, y_train)
print(f"{'VotingÂ (Hard)':>15} | Test: {accuracy_score(y_test, voting.predict(X_test)):.3f}")
# --------------------------------------------------------------------------- #
```
